```python
import pickle
import sys

sys.path.append('../../../Parsing/')
from language_elements import Sentence, Chunk, Word

with open('../../../Parsing/full_dataset_131.pickle', 'rb') as f2:
    retrieved_sentences = pickle.load(f2)
```

```python
X_train = []
Y_train = []

words = []
tags = []

word_count = dict()
tag_count = dict()

count = 0
for sentence in retrieved_sentences:
    tempX = []
    tempY = []
    for chunk in sentence.list_of_chunks:
        for word in chunk.list_of_words:
            words.append(word.kannada_word)
            tags.append(word.pos)
            
#             print("Word ", word.kannada_word)
#             print("tags ", word.pos)
            word_count[word.kannada_word] = word_count.get(word.kannada_word, 0)+1
            tag_count[word.pos] = tag_count.get(word.pos, 0)+1
            count += 1
            

            tempX.append(word.kannada_word)
            tempY.append(word.pos)
    
    X_train.append(tempX)
    Y_train.append(tempY)
    
for tag, cnt in tag_count.items():
    print(tag, cnt, sep='\t\t')
    
print(len(tag_count.keys()))
```
